```python
from matplotlib import pyplot
import pandas as pd
from pypfopt.expected_returns import mean_historical_return
from pypfopt.risk_models import sample_cov


def reader():
    return pd.read_csv('stock4.csv', delimiter=',', header=0)

def plot_creator(): # 1 task
    pyplot.rcParams['figure.autolayout'] = True
    reader().set_index('Date').plot()
    pyplot.show() # show time graph

def daily_income(): # 2 task
    headers = ['Apple', 'Cisco', 'IBM', 'Microsoft']
    dividends = [0.23, 0.38, 1.64, 0.62]
    for i in range(1, 5):
        div = dividends[i-1]
        price_prev = float(reader().iat[int(len(reader()) - 2), i])
        price_act = float(reader().iat[int(len(reader()) - 1), i])
        income = round(((price_act + div - price_prev) / price_prev) * 100, 2)
        result = headers[i-1] + ' income for ' + str(reader().iat[int(len(reader()) - 1), 0]) + ': ' + str(income) + '%'
        print(result)

def historical_return(): # 9 task a
    df = pd.read_csv('stock4.csv', parse_dates=True, index_col='Date')
    mu = mean_historical_return(df)
    S = sample_cov(df)
    print(str(mu) + '\n\n' + str(S))
```

```python
plot_creator()
```
